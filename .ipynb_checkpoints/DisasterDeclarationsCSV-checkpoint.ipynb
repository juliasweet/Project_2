# Data Cleaning 2

### Converting FemaWebDisasterDeclarations.csv into DisasterDeclarations.csv for use in our database

```python
import pandas as pd
```

```python
dirty_declarations_table = pd.read_csv("FemaWebDisasterDeclarations.csv")
```

```python
dirty_declarations_table.head()
```

### Our PostgreSQL table does not need columns that reference database maintenance/recordkeeping. Create a new table that uses only the columns that we will have in our PostgreSQL table. 

```python
disasterDeclarations = dirty_declarations_table[['disasterNumber', 'declarationDate', 'disasterName', 'incidentBeginDate', 'incidentEndDate', 'declarationType', 'stateCode', 'stateName', 'incidentType']].copy()
```

```python
disasterDeclarations.head()
```

```python
# Check data types
print(disasterDeclarations.dtypes)
```

### Remove hour:minute:second:timezone

```python
disasterDeclarations['declarationDate'] = disasterDeclarations['declarationDate'].str.replace('T00:00:00.000Z', '')
```

```python
disasterDeclarations['incidentBeginDate'] = disasterDeclarations['incidentBeginDate'].str.replace('T00:00:00.000Z', '')
```

```python
disasterDeclarations['incidentEndDate'] = disasterDeclarations['incidentEndDate'].str.replace('T00:00:00.000Z', '')
```

```python
disasterDeclarations.head()
```

```python
disasterDeclarations.to_csv("DisasterDeclarations.csv")
```

```python
disasterDeclarations = disasterDeclarations.rename(columns={'disasterName': 'disastername', 'incidentBeginDate': 'incidentbegindate',
                                                           'incidentEndDate': 'incidentenddate','declarationType': 'declarationtype',
                                                            'stateCode': 'statecode', 'stateName':'statename', 'incidentType':'incidenttype',
                                                            'disasterNumber':'disasternumber','declarationDate':'declarationdate'
                                                        })
```

```python
disasterDeclarations.head()
```

# Export to PostgreSQL

```python
from sqlalchemy import create_engine
```

```python
connection_string = "postgres:postgres@localhost:5432/project_2"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
disasterDeclarations.to_sql(name='disasterdeclarationstwo', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from disasterdeclarationstwo', con=engine).head()
```
