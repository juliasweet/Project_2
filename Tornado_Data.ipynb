# Query Only Tornadoes 

Use Pandas to get data only on tornadoes to compare with PostgreSQL query.

```python
import pandas as pd
```

```python
# Set file paths. 
summaries_file = 'DisasterSummaries.csv'
declarations_file = 'DisasterDeclarations.csv'

# Read CSVs
disaster_summaries_df = pd.read_csv(summaries_file)
disaster_declarations_df = pd.read_csv(declarations_file)
```

```python
disaster_summaries_df["totalFEMAapproved"] = disaster_summaries_df["totalNumberIaApproved"] + disaster_summaries_df["totalAmountIhpApproved"]
```

```python
disaster_declarations_df = disaster_declarations_df.loc[disaster_declarations_df["incidentType"]=="Tornado", :]
```

```python
disaster_declarations_df.head()
```

```python
combined_data = pd.merge(
disaster_declarations_df[['disasterNumber', 'declarationDate', 'incidentType', 'stateCode']],
disaster_summaries_df[['disasterNumber', 'totalNumberIaApproved', 'totalAmountIhpApproved']],
on ='disasterNumber',
how='outer')
```

```python
combined_data
```

```python
print(combined_data.columns)
```

```python
combined_data["declarationDate"]=combined_data["declarationDate"].str.replace('T00:00:00.000Z', '')
```

```python
combined_data.head()
```

```python
combined_data = combined_data.rename(columns={'disasterNumber':'disasternumber',
                                             'totalNumberIaApproved':'totalnumberiaapproved',
                                             'totalAmountIhpApproved':'totalamountihpapproved',
                                             'incidentType':'incidenttype',
                                             'declarationDate': 'declarationdate',
                                             'disasterName': 'disastername',
                                             'stateCode':'statecode'})
```

```python
combined_data.head()
```

```python
combined_data.to_csv('tornado_data.csv')
```
